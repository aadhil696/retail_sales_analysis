```python
import pandas as pd
data = pd.read_csv('orders.csv')
data.head(20)
```

### Data Cleaning


```python
#Check unique values in the column to see the presence of different null values
data['Ship Mode'].unique()
```

```python
#Change different unknown values as null valuess
data = pd.read_csv('orders.csv',na_values=['Not Available', 'unknown'])
data['Ship Mode'].unique()
```

```python
#Standardize columns by lower casing first letter and replace space with '_'
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(' ','_')
data.head(5)
```

```python
#check the data types and convert it if necessary
data.dtypes
```

```python
#change order_date dtype from object to date
data['order_date'] = pd.to_datetime(data['order_date'], format = "%Y-%m-%d")
data.dtypes
```

### Feature Engineering - Creating Business Metrics

```python
#find discount amount with list price and discount percentage
data['discount'] = data['list_price'] * data['discount_percent'] * .01
data.head(5)
```

```python
#find sale price with list price and discount amount
data['sale_price'] = data['list_price'] - data['discount']
data.head(5)
```

```python
#find the profit
data['profit'] = data['sale_price'] - data['cost_price']
data.head(5)
```

```python
#remove unnecessary columns
data.drop(columns = ['list_price', 'cost_price', 'discount_percent'], inplace = True)
data.head(5)
```

### Oracle Database Setup & Data Integration

```python
import sqlalchemy as sa

#Oracle details
username = "system"
password = "123123"
host = "192.168.1.8"
port = "1521"
service_name = "XE"

#Create connection string
dsn = sa.engine.URL.create(
    "oracle+oracledb",
    username=username,
    password=password,
    host=host,
    port=port,
    database=service_name
)

engine = sa.create_engine(dsn)

#Test connection
with engine.connect() as conn:
    print("COnnected:", conn)
```

```python
from sqlalchemy import Numeric
dtype_mapping = {col: Numeric() for col in data.select_dtypes("float").columns} #to overcome oracle float-binary error
data.to_sql('df_orders',con = engine, index = False, if_exists = 'replace', dtype=dtype_mapping)
```
